```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from IPython.display import clear_output
```

```python
url = 'https://www.flipkart.com/search?q=macbook&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
req = requests.get(url)
print(req)
soup = BeautifulSoup(req.text,'html')
print(soup.prettify())
```

```python
ty_mac=soup.findAll('div',attrs={'class':'col col-7-12'})  #mac types
ty_mac=[i.text for i in ty_mac]
print(len(ty_mac))
```

```python
s=soup.findAll('div',attrs={'class':'gUuXy-'})                       #star rating
star=[i.span.text for i in s]
print(star)
len(star)
```

```python
r=soup.findAll('span',attrs={'class':'_2_R_DZ'})
for i in r :
  print(i.span.span.text)
```

```python
pr_mac=soup.findAll('div',attrs={'class':'_30jeq3 _1_WHN1'}) #prices
pr_mac=[i.text for i in pr_mac]
print(len(pr_mac))
```

```python
rev_mac=soup.findAll('span',attrs={'class':'_2_R_DZ'}) #reviews
for i in rev_mac:
  print(i.text)
print(len(rev_mac))
```

```python
ratings=[]
reviews=[]
r=soup.findAll('span',attrs={'class':'_2_R_DZ'})
for i in r:
   r=i.text.split('&')
   ratings.append(r[0].split(' ')[0])
   reviews.append(r[1].split(' ')[0].strip())
print(ratings)
print(reviews)

len(ratings)
len(reviews)
```

```python
specification=[]
spec_mac=soup.findAll('ul',attrs={'class':'_1xgFaf'})  #specifications
for i in spec_mac:
  spec_mac=i.findAll('li')
  specification.append([i.text for i in spec_mac])
  
specification
len(specification)
```

```python
data=pd.DataFrame({'Product_info': ty_mac,
                   'Price': pr_mac,
                   'stars_info': star,
                   'Ratings': ratings,
                   'Reviews': reviews,
                   'Specifications': specification})
data
```

```python
data.to_csv('Macbook.csv',index=False)
```

```python
import time
base_url='https://www.flipkart.com/search?q=macbook&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_HistoryAutoSuggest_1_2_na_na_na&otracker1=AS_QueryStore_HistoryAutoSuggest_1_2_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=macbook%7CLaptops&requestId=fac3745a-260d-4e22-8a62-d7000b982913&page='

product=[]
price=[]
star=[]
reviews=[]
ratings=[]
specifications=[]


for i in range(1,3):
  time.sleep(3)
  url=base_url+str(i)
  req=requests.get(url)
  soup=BeautifulSoup(req.text,'html')
  
  ty_mac=soup.findAll('div',attrs={'class':'_4rR01T'})                #product name
  for i in ty_mac:
    product.append(i.text)
    pr_mac=soup.findAll('div',attrs={'class':'_30jeq3 _1_WHN1'})           # price
  for i in pr_mac:
    price.append(i.text)
    s=soup.findAll('div',attrs={'class':'gUuXy-'})                    #star rating
  for i in s:
    star.append(i.text)
    r=soup.findAll('span',attrs={'class':'_2_R_DZ'})       #rating n reviews
  for i in r:
    r=i.text.split('&')
    ratings.append(r[0].split(' ')[0])
    reviews.append(r[1].split(' ')[0].strip())
    spec_mac=soup.findAll('ul',attrs={'class':'_1xgFaf'})                #features
  for i in spec_mac:
    spec_mac=i.findAll('li')
    specifications.append([i.text for i in spec_mac])
```

```python
data=pd.DataFrame({'Product_info': product,
                   'Price': price,
                   'stars_info': star,
                   'Ratings': ratings,
                   'Reviews': reviews,
                   'Specifications': specifications})
data
```

```python

data.to_csv('Macbook.csv',index=False)
```
